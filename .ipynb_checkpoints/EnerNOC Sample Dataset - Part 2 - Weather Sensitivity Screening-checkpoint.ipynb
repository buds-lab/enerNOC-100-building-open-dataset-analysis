##Part 1 - Weather Sensitivity Screening

This notebook takes the [EnerNOC](http://open.enernoc.com/data/) dataset of 100 anonymized buildings for the 2012 year and applies the weather sensitivity screening approach.

Data stream influence characterization is the process of roughly classifying the dataset into streams and subsequences based on weather conditions sensitivity. This evaluation is important in understanding what measured performance is due to heating, cooling, and ventilation systems (HVAC) responses to outdoor conditions and what is due to schedule, occupancy, lighting, and miscellaneous loading conditions which are weather independent. Performance data that is influenced by weather can be used to better understand the HVAC system operation or be weather-normalized to understand occupant diversity schedules. Non-weather sensitive data streams are used with less pre-processing to create diversity schedules and to calculate miscellaneous and lighting load power densities.

This [dataset](http://open.enernoc.com/data) and is licensed under a [Creative Commons Attribution-NonCommercial 3.0 Unported License](http://creativecommons.org/licenses/by-nc/3.0/).

##Load all the buildings and combine with weather data

First, we create an empty Python `dict` object, name each item according to the `SITE_ID`, and use the [`.concat()`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.tools.merge.concat.html) funtion to combine all the files and we'll iterate through the input files, load each one, and merge them into this frame.

For each file, we'll load the weather file as well.

```python
rawfilepath = "./all-data/csv/"
```

```python
filelist = os.listdir(rawfilepath)
```

```python
weatherfiledir = "./wunderground/"
```

```python
all_raw_dict = {}

for filename in filelist:
    
    try:
        df = pd.read_csv(os.path.join(rawfilepath,filename), index_col="dttm_utc", parse_dates=True)
        df = df.drop(["timestamp", "estimated", "anomaly"], axis=1).tz_localize('UTC')
        df.columns = ["EnergyCons_kWh"]
        
        #Resample mean to hourly to get kW avg across hour -- which is can be used as an effective kWh
        df = df.resample('H')
        
        #Then resample by summing across days
        df = df.resample('D', how='sum')

        #Get the SITE_ID from the filename
        SITE_ID = filename[:-4]

        weatherfilename = meta_weatherfile[(meta_weatherfile["SITE_ID.1"] == float(SITE_ID))]["Weather File"].values
        weatherdata = pd.read_csv(weatherfiledir+weatherfilename[0]+".csv", index_col=[0], parse_dates=True) 

        df_merged = pd.merge(df, weatherdata.tz_localize("UTC"), right_index=True, left_index=True)

        all_raw_dict[SITE_ID] = df_merged

    except:        
        print "Error loading "+filename 
```

```python
all_raw_df = pd.concat(all_raw_dict, axis=1)
```

Now we have a DataFrame with each location's daily energy consumption and various weather measurements

```python
all_raw_df.head()
```

##Calculate Weather Sensitity using Spearman Rank Coefficient

In this filtering step we use the Spearman Rank Order Correlation (ROC) to evaluate the positive or negative correlation between each performance measurement point and the outdoor air dry bulb temperature. This technique has been previously used for [weather sensitivity analysis](http://drrc.lbl.gov/publications/statistical-analysis-baseline-load). The ROC coefficient, $\rho$, is calculated according to a comparison of two data streams, $X$ and $Y$, in which the values at each time step, $X_i$ and $Y_i$, are converted to a relative rank of magnitude, $x_i$ and $y_i$, according to its respective dataset. These rankings are then used to calculate $\rho$ that varies between +1 and -1 with each extreme corresponding to a perfect positive and negative correlation respectively. A value of 0 signifies no correlation between the datasets. This $\rho$ value for a time-series is calculated according the following equation.

$\rho = 1 - \frac{6\sum d_i^2}{n(n^2-1)}$

The difference between the point rankings, $x_i$ and $y_i$, is signified by a difference value, $d_i$, and the number of samples compared in each dataset is signified by $n$. 

```python
from IPython.display import Image
Image("./graphics/weathersensitivity_examples.png")
```

The Figure above illustrates the calculation of the ROC coefficient, $\rho$ for three examples. The cooling sensitive data set shows a strong positive correlation between outside air temp and energy consumption with a $\rho$ value of 0.934. As the outside air temperature increases, the power consumption measured by this meter increases. The heating sensitive dataset shown has a strong negative correlation with a $\rho$ of -0.68. A weather insensitive dataset is shown in the middle which has a $\rho$ of 0.0, signifying no weather correlation, which is obvious due to the four levels of consumption which are independent from outdoor air conditions.

##Calculate the Spearman Rank Correlation for performance for all locations and visualize as a comparison

Let's cycle through all the locations and find the weather sensitivity for each. First, we need to load the `scipy.stats` library of functions which includes the [`spearmanr`](https://docs.scipy.org/doc/scipy-0.7.x/reference/generated/scipy.stats.spearmanr.html) function.

```python
import scipy.stats as stats
```

Let's look at the $\rho$ value for all of the locations. We will calculate the $\rho$ value for each month individually to understand how spearman changes over time.

```python
spearmans = pd.DataFrame()

for datapoint in list(all_raw_df.columns.levels[0]):
    try:
        combined = pd.merge(pd.DataFrame(all_raw_df[datapoint]['EnergyCons_kWh']), pd.DataFrame(all_raw_df[datapoint]['Max TemperatureC']), right_index=True, left_index=True, how='left')
        combined = combined.dropna()

        spearman = combined.groupby([lambda x: x.year, lambda x: x.month]).apply(lambda x: stats.spearmanr(x['EnergyCons_kWh'], x['Max TemperatureC'])[0])
        spearman = spearman.reset_index()
        spearman.index = pd.to_datetime((spearman.level_0*100 + spearman.level_1).apply(str), format='%Y%m')
        spearmans[datapoint] = spearman[0]
#         print "Finished loading "+datapoint
    except:
        print datapoint+" unsuccessful"
```

Now we have a dataframe with coefficients for each month
